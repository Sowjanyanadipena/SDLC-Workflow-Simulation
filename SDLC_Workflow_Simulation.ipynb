```python
# SDLC Workflow Simulation Project

project = {
    "requirements": None,
    "design": None,
    "development": None,
    "testing": None,
    "deployment": None,
    "maintenance": None
}

def enter_requirements():
    project["requirements"] = input("Enter project requirements: ")
    print("\nRequirements added successfully!")

def design_phase():
    if not project["requirements"]:
        print("\nAdd requirements first!")
        return
    project["design"] = input("Enter system design details: ")
    print("\nDesign phase completed successfully!")

def development_phase():
    if not project["design"]:
        print("\nComplete design phase first!")
        return
    project["development"] = "Development completed based on design."
    print("\nDevelopment phase completed successfully!")

def testing_phase():
    if not project["development"]:
        print("\nComplete development phase first!")
        return
    bugs = input("Enter any bugs found during testing (or 'None'): ")
    project["testing"] = f"Testing done. Bugs: {bugs}"
    print("\nTesting phase completed successfully!")

def deployment_phase():
    if not project["testing"]:
        print("\nComplete testing phase first!")
        return
    project["deployment"] = "Project deployed successfully!"
    print("\nDeployment phase completed successfully!")

def maintenance_phase():
    if not project["deployment"]:
        print("\nComplete deployment first!")
        return
    updates = input("Enter maintenance updates: ")
    project["maintenance"] = f"Maintenance logs: {updates}"
    print("\nMaintenance phase completed successfully!")

def view_project_status():
    print("\n--- Project Status ---")
    for phase, detail in project.items():
        print(f"{phase.capitalize()}: {detail if detail else 'Pending'}")

# --- Main Menu ---
while True:
    print("\n=== SDLC Workflow Simulation ===")
    print("1. Enter Requirements")
    print("2. Design Phase")
    print("3. Development Phase")
    print("4. Testing Phase")
    print("5. Deployment Phase")
    print("6. Maintenance Phase")
    print("7. View Project Status")
    print("8. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        enter_requirements()
    elif choice == "2":
        design_phase()
    elif choice == "3":
        development_phase()
    elif choice == "4":
        testing_phase()
    elif choice == "5":
        deployment_phase()
    elif choice == "6":
        maintenance_phase()
    elif choice == "7":
        view_project_status()
    elif choice == "8":
        print("Exiting SDLC Simulation. Goodbye!")
        break
    else:
        print("Invalid choice! Please try again.")
```
